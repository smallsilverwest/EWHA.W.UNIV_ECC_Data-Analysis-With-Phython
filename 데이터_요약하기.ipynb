```python
import numpy as np
import pandas as pd
```

```python
a=list(range(1,11))
print(a)
```

```python
#평균 구하기
np.mean(a)
```

```python
#중앙값 구하기
np.median(a)
```

```python
#분산 구하기
np.var(a)
```

```python
#비편향 분산
np.var(a, ddof=1)
```

```python
#standard deviation 구하기
np.std(a),
```

```python
#percentile 구하기에서 오류가 발생하는 경우
np.percentile(a)
```

```python
#percentile을 구하기 위해서는 괄호 내, 분위수 %를 설정해야함
np.percentile(a,0)
```

```python
np.percentile(a,25)
```

```python
np.percentile(a,75)
```

```python
pd.Series(a).quantile(0.9, interpolation='midpoint')
```

```python
pd.Series(a).quantile(0.95, interpolation='nearest')
```

```python
b = np.array([18,   5,  10,  23,  19,  -8,  10,   0,   0,   5,   2,  15,   8,
              2,   5,   4,  15,  -1,   4,  -7, -24,   7,   9,  -6,  23, -13])
```

```python
#데이터의 개수
len(b)
```

```python
#최댓값과 최솟값
np.max(b)
np.min(b)
```

```python
print(np.percentile(b, 0))
print(np.percentile(b, 25))
print(np.percentile(b, 50))
print(np.percentile(b, 75))
print(np.percentile(b, 100))
```

```python
import gdown
gdown.download('https://bit.ly/3pK7iuu','ns_book7.csv',quiet=False)
```

```python
ns_book7=pd.read_csv('ns_book7.csv',low_memory=False)
ns_book7.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#산점도 그리기
plt.scatter([1,2,3,4],[1,2,3,4])
plt.show()
```

```python
plt.scatter(ns_book7['번호'],ns_book7['대출건수'])
plt.show()
```

```python
#투명도 조절하기
plt.scatter(ns_book7['번호'],ns_book7['대출건수'],alpha=0.1)
plt.show()
```

```python
average_borrows=ns_book7['대출건수']/ns_book7['도서권수']
plt.scatter(average_borrows, ns_book7['대출건수'],alpha=0.1)
plt.show()
```

```python
#히스토그램 그리기
plt.hist([0,3,5,6,7,7,8,13],bins=5)
plt.show()
```

```python
np.histogram_bin_edges([0,3,5,6,7,7,9,13],bins=5)
```

```python
plt.hist(ns_book7['대출건수'])
plt.show()
```

```python
plt.hist(ns_book7['대출건수'])
plt.yscale('log')
plt.show()
```

```python
title_len=ns_book7['도서명'].apply(len)
plt.hist(title_len,bins=100)
plt.show()
```

```python
#boxplot 그리기
plt.boxplot(ns_book7[['대출건수','도서권수']])
plt.yscale('log')
plt.show()
```

```python
plt.boxplot(ns_book7[['대출건수','도서권수']], vert=False)
plt.xscale('log')
plt.show()
```

```python
plt.boxplot(ns_book7[['대출건수','도서권수']], whis=(0,100))
plt.yscale('log')
plt.show()
```
