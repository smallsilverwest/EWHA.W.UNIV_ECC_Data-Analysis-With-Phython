```python
#뷰티플수프로 누라된 값 채우기

import requests
from bs4 import BeautifulSoup
```

```python
def get_book_title(isbn):
  url='https://www.yes24.com/Product/Goods/96024871'
  r=requests.get(url.format(isbn))
  soup=BeautifulSoup(r.text, 'html.parser')

  title=soup.find('a', attrs={'class':'gd_name'})\
  .get_text()

  return title
```

```python
#9791162243664은 '혼자 공부하는 머신러닝+딥러닝' ISBM13

get_book_title(9791162243664)
```
