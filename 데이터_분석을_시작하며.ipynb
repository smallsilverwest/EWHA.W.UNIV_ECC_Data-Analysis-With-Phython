```python
#Python open()함수는 텍스트 파일을 UTF-8 형식으로 가정
#해당 파일이 완성형 인코딩 EUC-KR을 사용해 오류가 발생하는 경우

with open('남산도서관 장서 대출목록 (2021년 04월).csv') as f:
  print(f.readline())
```

```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', mode='rb') as f:
  d = f.readline()
print(chardet.detect(d))
```
